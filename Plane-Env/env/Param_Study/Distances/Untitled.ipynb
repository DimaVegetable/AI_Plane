```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
os.listdir()
%matplotlib inline
```

```python
list_distances = []
list_params = []
for i in range(1,82):
    new_path = os.path.join(str(i),"distances.txt")
    text_file = open(new_path, "r")
    list_distances.append(eval(text_file.read()))
    new_path = os.path.join(str(i),"distances.txt")
    text_file = open(new_path, "r")
    list_params.append(eval(text_file.read()))
    
```

```python
fig, ax = plt.subplots()
for i in range(0,len(list_distances)):
    pd.Series(list_distances[i]).plot(figsize=((20,20)),ax=ax)
```

```python
df = pd.DataFrame(list_distances).transpose()
```

```python
df.mean().sort_values(ascending=False).reset_index(drop=True).plot()
```

```python
df.std().sort_values(ascending=False).reset_index(drop=True).plot()
```

```python
df.min().plot()
```
